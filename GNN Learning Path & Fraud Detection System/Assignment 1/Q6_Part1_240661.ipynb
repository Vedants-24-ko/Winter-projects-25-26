```python
import matplotlib.pyplot as plt

# creating python files manually
square_feet = [
    1100,1400,1425,1550,1600,1700,1750,1800,1875,2000,
    2100,2250,2300,2400,2450,2600,2800,2900,3000,3150,3300
]

prices = [
    199000,245000,230000,215000,280000,295000,345000,315000,325000,360000,
    350000,385000,390000,425000,415000,455000,465000,495000,510000,545000,570000
]
```

```python


n = len(square_feet)

sum_x = sum(square_feet)
sum_y = sum(prices)
sum_xy = sum(square_feet[i] * prices[i] for i in range(n))
sum_x2 = sum(x*x for x in square_feet)

m_ols = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - (sum_x ** 2))
b_ols = (sum_y - m_ols * sum_x) / n

print("OLS Slope (m):", m_ols)
print("OLS Intercept (b):", b_ols)
```

```python
x_test = 2500
pred_ols = m_ols * x_test + b_ols

print(f"Predicted price for {x_test} sq ft (OLS): ${pred_ols:,.2f}")
```

```python
plt.figure(figsize=(8,5))
plt.scatter(square_feet, prices, label="Data Points")
plt.plot(square_feet, [m_ols * x + b_ols for x in square_feet],
         color='red', label="OLS Best Fit Line")

plt.xlabel("Square Footage")
plt.ylabel("Price")
plt.title("OLS Line of Best Fit")
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```
